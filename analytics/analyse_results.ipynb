# Analysing the Game Statistics of many Simulated Rounds 'The Game'

The player strategies so far are pretty basic. These are the first two results for playing a maximum of two cards and â€¦
1. Minimizing the overall Top Card Score (see definition below)
2. The above strategy + player communication when a player can reverse a stack with his/her hand cards

### Top Card Score
$TCS = \sum ^{IS} _{is} top_{is} - 1 + \sum ^{DS} _{ds} 100 - top_{ds}$,

where $IS$ are the two increasing stacks and $DS$ are the two decreasings stacks with their current top card on the deck $is$ and $ds$ respectively.
* The game starts with a $TCS$ of 0 (i.e. top cards beeing two times 1 and two times 100).
* The $TCS$ at the end the game can lie between 98 (best) and 392 (worst).

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_theme(context="poster", palette="Paired", font_scale=0.8)
import os, sys
```

## 1. Without Communication (minimize top card score)

```python
# Load data

wins = np.fromfile("../wins_01.txt", dtype=int, sep=",")
losses = np.fromfile("../losses_01.txt", dtype=int, sep=",")
print(f"Read {len(wins)} win scores and {len(losses)} remaining card numbers for lost games.")
```

```python
plt.figure(figsize=(12,8))
plt.title("How often is the game finished without communication?\n (based on 1000 simulated games)")
plt.pie([len(wins), len(losses)], explode=[0.1,0.1], autopct=lambda pct: f"{pct:.2f} %")
plt.legend(labels=["wins", "defeats"], loc="upper right")
plt.show()
```

```python
plt.figure(figsize=(16,8))
plt.title("How many cards are remaining when losing the game?")
ax = sns.histplot(losses, bins=range(2,50))
plt.show()
```
